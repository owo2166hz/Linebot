介紹
每當您在 Python 中建立變數時，它都會有一個具有對應資料類型的值。 有許多不同的資料類型，例如整數、浮點數、布林值和字串，我們將在本課中介紹所有這些類型。 （這只是可用資料類型的一小部分——還有字典、集合、列表、元組等等。）

資料類型很重要，因為它們決定了您可以使用它們執行哪些操作。 例如，您可以劃分兩個浮點數，但不能分割兩個字串。 例如，12.0/2.0 有意義，但「cat」/「dog」則沒有意義。

為了避免錯誤，我們需要確保操作與我們擁有的資料類型相符。

整數
整數是沒有任何小數部分的數字，可以是正數 (1, 2, 3, ...)、負數 (-1, -2, -3, ...) 或零 (0)。

在下面的程式碼單元中，我們將變數 x 設定為整數。 然後我們使用 type() 驗證資料類型，並且只需將變數名稱傳遞到括號中。



x = 14
列印（x）
列印（類型（x））
14
<類'int'>
在上面的輸出中，<class 'int'> 指的是整數資料型態。

近_pi = 3.141592653589793238462643383279502884197169399375105820974944

幾乎_pi = 22/7

# 四捨五入至小數點後 5 位
rounded_pi = 圓(almost_pi, 5)

每當你寫一個帶有小數點的數字時，Python 都會將其識別為浮點資料類型。

例如，1.（或 1.0、1.00 等）將被辨識為浮點數。 情況確實如此，儘管這些數字在技術上沒有小數部分！

y_浮點 = 1。
列印（y_float）
列印（類型（y_float））
1.0
<類'浮動'>
布林值¶
布林值表示兩個值之一：True 或 False。 在下面的程式碼單元中，z_one 設定為值為 True 的布林值。

z_one = 真
列印（z_one）
列印（類型（z_one））
真的
<類'布林'>
接下來，z_two 設定為值為 False 的布林值。

z_two = 假
列印（z_two）
列印（類型（z_two））
錯誤的
<類'布林'>
布林值用於表示表達式的真值。 由於 1 < 2 是 true 語句，因此 z_third 的值為 True。

z_三 = (1 < 2)
列印（z_三）
列印（類型（z_三））
真的
<類'布林'>
類似地，由於 5 < 3 是錯誤陳述，因此 z_four 的值為 False。

z_four = (5 < 3)
列印（z_four）
列印（類型（z_four））
錯誤的
<類'布林'>
我們可以使用 not 來切換布林值。 因此，not True 相當於 False，not False 則變成 True。

z_ Five = 不是 z_four
列印（z_ Five）
列印（類型（z_ Five））
真的
<類'布林'>
當您學習條件和條件語句時，布林值在下一課中將很重要。

弦樂
字串資料類型是包含在引號中的字元（如字母、標點符號、數字或符號）的集合。 字串通常用於表示文字。

w =“你好，Python！”
列印（w）
列印（類型（w））
你好，Python！
<類'str'>
您可以使用 len() 來取得字串的長度。 “你好，Python！” 長度為 14，因為它有 14 個字符，包括空格、逗號和感嘆號。 請注意，計算長度時不包括引號。

列印（長度（w））
14
一種特殊類型的字串是空字串，其長度為零。

最短字串=“”
列印（類型（最短字串））
列印（len（最短的字串））
<類'str'>
0
如果將數字放在引號中，則它具有字串資料類型。

我的號碼 =“1.12321”
列印（我的號碼）
列印（類型（我的號碼））
1.12321
<類'str'>
如果我們有一個可轉換為浮點數的字串，我們可以使用 float()。

這並不總是有效！ 例如，我們可以將「10.43430」和「3」轉換為浮點數，但不能將「Hello, Python!」轉換為浮點數。 到一個浮動。

也我的號碼 = 浮點數（我的號碼）
列印（也是我的號碼）
列印（類型（also_my_number））
1.12321
<類'浮動'>
就像您可以添加兩個數字（浮點數或整數）一樣，您也可以添加兩個字串。 它會產生一個更長的字串，透過連接兩個原始字串來組合它們。

new_string =“abc”+“def”
列印（新字串）
列印（類型（新字串））
abcdef
<類'str'>
請注意，不可能對兩個字串進行減法或除法。 您也不能將兩個字串相乘，但可以將一個字串與一個整數相乘。 這再次產生一個字串，該字串只是原始字串與其自身連接指定次數的字串。

最新字串=“abc”* 3
列印（最新的字串）
列印（類型（最新字串））
ABCABCABC
<類'str'>
請注意，您不能將字串與浮點數相乘！ 嘗試這樣做將傳回錯誤。

will_not_work =“abc”* 3。

類型錯誤：無法將序列乘以“float”類型的非 int
錯誤中，“sequence”是字串“abc”，“non-int of type 'float'”是float（3.）。 因此，錯誤訊息可以改寫為「無法將字串乘以浮點數」。
